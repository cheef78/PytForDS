Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
x = float(input("Введите координату X вектора = "))
y = float(input("Введите координату У вектора = "))
z = float(input("Введите координату Z вектора = "))
```

```python
print ('Длина вектора = ',round(m.sqrt(x*x+y*y+z*z),2))
```

Напишите код на Python, реализующий построение графиков:
окружности, эллипса, гиперболы.

```python
fi=np.linspace(0,360,360)
r = float(input("Введите значение радиуса окружности - R = "))
plt.plot(np.cos(fi)*r,np.sin(fi)*r)
plt.grid(color='lightgray',linestyle='--')
```

```python
a = float(input("Введите значение большой полуоси эллипса  = "))
b = float(input("Введите значение малой полуоси эллипса  = "))
plt.plot( a*np.cos(fi) , b*np.sin(fi) )
plt.grid(color='lightgray',linestyle='--')
```

```python
k = float(input("Введите значение параметра k  = "))
b = float(input("Введите значение смещения b  = "))
x1=np.linspace(0.1,2,1000)
x2=np.linspace(-0.1,-2,1000)
plt.plot( x1, (b+(k/x1)))
plt.plot( x2, (b+(k/x2)))
plt.grid(color='lightgray',linestyle='--')
```

```python
from mpl_toolkits .mplot3d import Axes3D
x1=np.linspace(-2,2,1000)
y1=np.linspace(-2,2,1000)
x2=np.linspace(-2,2,1000)
y2=np.linspace(-2, 2,1000)
xgrid1 , ygrid1 = np.meshgrid(1*x1, 1*y1)
zgrid1 = -(xgrid1+ygrid1-25)
xgrid2 , ygrid2 = np.meshgrid(1*x2, 1*y2)
zgrid2 = -(xgrid2+ygrid2)




frame = plt.figure(figsize =(10 ,8))
axes = Axes3D(frame)
axes. plot_surface (xgrid1 , ygrid1 , zgrid1 )
axes. plot_surface (xgrid2 , ygrid2 , zgrid2 )
plt.show ()
```
