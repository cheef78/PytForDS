
4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
x=np.linspace(-2,4.5,101)
y=np.exp(x)+2*x-x**3-1
plt.figure(figsize =(10 ,10))
plt.plot(x,y)
plt.grid(color='lightgray',linestyle='--')
plt.minorticks_on()
plt.grid(which='minor', color = 'k', linestyle = ':')
```

Из графика получили 4 корня -1.6; 0; 2.6; 4.2

```python
from scipy.optimize import fsolve
def equations(p):
    x,y = p
    return (x**2-1,m.exp(x)+x*(1-y)-1)

x0=np.array([4.52,0])
x,y=  fsolve(equations, x0)

print (x,y)
```
